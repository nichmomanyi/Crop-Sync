#### Parse and clean Data
* This is because the csv file columns have values in concatenated strings in lists or dictionaries. 

```python
import ast
import re

def parse_list(column_value):
    """Parse a string representation of a list into an actual list."""
    try:
        return ast.literal_eval(column_value)
    except (ValueError, SyntaxError):
        return []

def parse_dict(column_value):
    """Parse a string representation of a dictionary into an actual dictionary."""
    try:
        return ast.literal_eval(column_value)
    except (ValueError, SyntaxError):
        return {}

def split_concatenated_string(concatenated_string):
    """Split a concatenated string into a list of items based on capital letters."""
    if not concatenated_string:
        return []
    items = re.findall(r'[A-Z][a-zA-Z0-9\(\)]+', str(concatenated_string))
    return items

# Apply parsing functions to all relevant columns
recommendation_table['Varieties'] = recommendation_table['Varieties'].apply(
    lambda x: split_concatenated_string(str(x).strip("[]'"))
)
recommendation_table['Crops'] = recommendation_table['Crops'].apply(
    lambda x: split_concatenated_string(str(x).strip("[]'"))
)
recommendation_table['Towns'] = recommendation_table['Towns'].apply(
    lambda x: split_concatenated_string(str(x).strip("[]'"))
)
recommendation_table['Countries'] = recommendation_table['Countries'].apply(
    lambda x: split_concatenated_string(str(x).strip("[]'"))
)
recommendation_table['TPE Descriptions'] = recommendation_table['TPE Descriptions'].apply(
    lambda x: split_concatenated_string(str(x).strip("[]'"))
)
recommendation_table['Stressors'] = recommendation_table['Stressors'].apply(
    lambda x: split_concatenated_string(str(x).strip("[]'"))
)
recommendation_table['Key Traits'] = recommendation_table['Key Traits'].apply(parse_dict)
recommendation_table['Key TPEs'] = recommendation_table['Key TPEs'].apply(parse_dict)
recommendation_table['Key Conditions'] = recommendation_table['Key Conditions'].apply(parse_dict)
recommendation_table['Key Months'] = recommendation_table['Key Months'].apply(parse_dict)
recommendation_table['Numeric Summary'] = recommendation_table['Numeric Summary'].apply(parse_dict)

recommendation_table.head()
```

* The code above defines helper functions to parse and clean the data.
- The functions are applied to all relevant columns to convert concatenated strings into lists or dictionaries.

```python
from collections import defaultdict

# Create a dictionary to map each variety to its countries
variety_to_countries = defaultdict(set)

for index, row in recommendation_table.iterrows():
    varieties = row['Varieties']
    countries = row['Countries']
    for variety in varieties:
        for country in countries:
            variety_to_countries[variety].add(country)

# Display the mapping for a few varieties
print("Variety to Countries Mapping (sample):")
for variety, countries in list(variety_to_countries.items())[:5]:
    print(f"{variety}: {countries}")
```

#### Analyze Clusters by Country
- This function filters clusters that include a specific country and displays the varieties and other relevant information.
- It helps you understand which varieties are recommended for specific countries.

```python
def analyze_clusters_by_country(country_of_interest):
    clusters_for_country = recommendation_table[
        recommendation_table['Countries'].apply(lambda x: country_of_interest in x)
    ]

    if clusters_for_country.empty:
        print(f"\nNo clusters found for {country_of_interest}.")
    else:
        print(f"\nClusters for {country_of_interest}:")
        for index, row in clusters_for_country.iterrows():
            print(f"\nCluster {row['Cluster']}:")
            print(f"Varieties: {row['Varieties']}")
            print(f"TPE Descriptions: {row['TPE Descriptions']}")
            print(f"Stressors: {row['Stressors']}")

# Example: Analyze clusters for Kenya
analyze_clusters_by_country('Kenya')
```

#### Analyze Each Cluster
* Loop iterates through each row in the recommendation table and prints the characteristics of each cluster.
* Helps you understand the traits, environmental conditions, and locations associated with each cluster.


```python
# Loop through each cluster to display its characteristics
for index, row in recommendation_table.iterrows():
    cluster = row['Cluster']
    print(f"\nCluster {cluster} Analysis:")

    # Display numeric summary
    print("\nNumeric Summary:")
    print(row['Numeric Summary'])

    # Display varieties
    print("\nVarieties:")
    print(row['Varieties'])

    # Display crops
    print("\nCrops:")
    print(row['Crops'])

    # Display towns
    print("\nTowns:")
    print(row['Towns'])

    # Display countries
    print("\nCountries:")
    print(row['Countries'])

    # Display TPE descriptions
    print("\nTPE Descriptions:")
    print(row['TPE Descriptions'])

    # Display stressors
    print("\nStressors:")
    print(row['Stressors'])

    # Display key traits
    print("\nKey Traits:")
    print(row['Key Traits'])

    # Display key TPEs
    print("\nKey TPEs:")
    print(row['Key TPEs'])

    # Display key conditions
    print("\nKey Conditions:")
    print(row['Key Conditions'])

    # Display key months
    print("\nKey Months:")
    print(row['Key Months'])
```
