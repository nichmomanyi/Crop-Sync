### ML Data preprocessing
* At this stage, we are preparing the data for machine learning by cleaning and transforming it into a format that algorithms can understand. This involves removing unnecessary columns, converting categorical data into numerical format using one-hot encoding, and scaling numerical features (like temperature and rainfall) to ensure they are on a similar scale. This preparation is crucial for accurate and effective machine learning analysis, which will help us recommend the best crop varieties for specific environments.

### Encoding categorical variables

* One-hot encoding is a technique used to convert categorical data (like names, labels, or categories) into a numerical format that machine learning algorithms can process. It creates new binary columns for each category in a variable, where a value of 1 indicates the presence of that category, and 0 indicates its absence. This prevents the algorithm from misinterpreting categorical data as having an ordinal relationship (e.g., assuming "2" is greater than "1").
* pd.get_dummies(ml_data, columns=['trait_name', 'trait_type', 'tpe_name', 'condition', 'month']) converts all specified categorical columns into one-hot encoded binary columns.

```python
ml_data_encoded = pd.get_dummies(ml_data, columns=['trait_name', 'trait_type', 'tpe_name', 'condition', 'month', 'country', 'crop_name'])

# Display the encoded columns
trait_name_cols = [col for col in ml_data_encoded.columns if col.startswith('trait_name_')]
trait_type_cols = [col for col in ml_data_encoded.columns if col.startswith('trait_type_')]
tpe_name_cols = [col for col in ml_data_encoded.columns if col.startswith('tpe_name_')]
condition_cols = [col for col in ml_data_encoded.columns if col.startswith('condition_')]
month_cols = [col for col in ml_data_encoded.columns if col.startswith('month_')]
country_cols = [col for col in ml_data_encoded.columns if col.startswith('country_')]
crop_name_cols = [col for col in ml_data_encoded.columns if col.startswith('crop_name_')]

print("Encoded Trait Name Columns:", trait_name_cols[:5])
print("\nEncoded Trait Type Columns:", trait_type_cols[:5])
print("\nEncoded TPE Name Columns:", tpe_name_cols[:5])
print("\nEncoded Condition Columns:", condition_cols[:5])
print("\nEncoded Month Columns:", month_cols[:5])
print("\nEncoded Country Columns:", country_cols[:5])
print("\nEncoded Crop Name Columns:", crop_name_cols[:5])
```

##### Map One-Hot Encoded Country Columns Back to Original Country
- To facilitate group the varieties to their countries

```python
# Function to map one-hot encoded country columns back to the original country
def get_country(row):
    for col in country_cols:
        if row[col] == 1:
            return col.replace('country_', '')
    return None

# Apply the function to each row to get the country
ml_data_encoded['country'] = ml_data_encoded.apply(get_country, axis=1)

# Verify the country column
print("Country column added:")
print(ml_data_encoded[['variety_name', 'country']].head())
```

```python
# Step 3: Check for missing values and drop them if any
ml_data_encoded = ml_data_encoded.dropna()
```

#### Select features for clustering

```python
# Numeric features (excluding variety_id, location_id, and year as they are identifiers)
numeric_features = ['lat', 'lon', 'temperature', 'humidity', 'rainfall']

# Combine all numeric and one-hot encoded features
analysis_cols = (
    numeric_features +
    trait_name_cols +
    trait_type_cols +
    tpe_name_cols +
    condition_cols +
    month_cols +
    country_cols +
    crop_name_cols
)

# Select the relevant columns for clustering
X = ml_data_encoded[analysis_cols]
```

### Scaling numerical features
* Scaling numerical features (like temperature and rainfall) to ensure they are on a similar scale **(normalizing independent variables or features of data) - data normalization**.
* Standardization or Feature scaling is the process of transforming the data we are working with into a standard scale. This translates to subtracting the mean and dividing by the standard deviation. In this way, regardless of the data set, we will always obtain a distribution with a mean of zero and a standard deviation of one, which could easily be proven.
* **standardscaler()** is used to standardize numerical features by removing the mean and scaling to unit variance.
* fit_transform: This method computes the mean and standard deviation of each feature (fit) and then scales the data so that each feature has a mean of 0 and a standard deviation of 1 (transform).
* The scaled values replace the original values in the ml_data_encoded DataFrame.

<h4>Why Scaling is Important: </h4>

* Algorithm Performance: Many machine learning algorithms (e.g., K-Means, SVM, Neural Networks) perform better when numerical features are on a similar scale. Scaling ensures that no single feature dominates the others due to its scale.
* Gradient Descent Convergence: Algorithms that use gradient descent (e.g., linear regression, logistic regression) converge faster when features are scaled.
* Distance Calculations: Algorithms that rely on distance calculations (e.g., K-Means, K-Nearest Neighbors) can be significantly affected by the scale of the features. Scaling ensures that each feature contributes equally to the distance calculation.
* **Scaling ensures that all numerical features contribute equally to the clustering process.**






```python
from sklearn.preprocessing import StandardScaler

# Initialize the StandardScaler
scaler = StandardScaler()

# Scale the numerical features
X[numeric_features] = scaler.fit_transform(X[numeric_features])

# Display the first few rows of the scaled data
print("First few rows of the scaled numerical features:")
X[numeric_features].head()
```

### Clustering with K-Means
**Objective:**
Group crop varieties based on their traits and environmental conditions to identify patterns and similarities.

**Explanation:** K-Means Clustering: This algorithm groups data into k clusters by minimizing the variance within each cluster.</br>
Why Use K-Means?: It is simple, efficient, and effective for grouping data points based on feature similarity.


#### Applying k-means to the clustered data

```python
from sklearn.cluster import KMeans

# Initialize K-Means with a chosen number of clusters
optimal_clusters = 5  # You can adjust this number based on domain knowledge or previous analysis
kmeans = KMeans(n_clusters=optimal_clusters, random_state=42)

# Fit the K-Means model to the data
kmeans.fit(X)

# Add cluster labels to the dataset
ml_data_encoded['cluster'] = kmeans.labels_

# Check the cluster assignments
print("Cluster assignments for each variety:")
cluster_assignments = ml_data_encoded[['variety_name', 'country', 'cluster']].drop_duplicates()
print(cluster_assignments)
```

#### Apply K-Means Clustering

#### Numeric and Categorical Columns

#### Analyzing Clusters

```python
# Numeric features (excluding variety_id, location_id, and year as they are identifiers)
numeric_features = ['lat', 'lon', 'temperature', 'humidity', 'rainfall']

# One-hot encoded features
trait_name_cols = [col for col in ml_data_encoded.columns if col.startswith('trait_name_')]
trait_type_cols = [col for col in ml_data_encoded.columns if col.startswith('trait_type_')]
tpe_name_cols = [col for col in ml_data_encoded.columns if col.startswith('tpe_name_')]
condition_cols = [col for col in ml_data_encoded.columns if col.startswith('condition_')]
month_cols = [col for col in ml_data_encoded.columns if col.startswith('month_')]

# Include one-hot encoded country and crop_name columns
country_cols = [col for col in ml_data_encoded.columns if col.startswith('country_')]
crop_name_cols = [col for col in ml_data_encoded.columns if col.startswith('crop_name_')]

# Combine all numeric and one-hot encoded features
analysis_cols = (
    numeric_features +
    trait_name_cols +
    trait_type_cols +
    tpe_name_cols +
    condition_cols +
    month_cols +
    country_cols +
    crop_name_cols
)

# Select the relevant columns for clustering
X = ml_data_encoded[analysis_cols]

# Display the first few rows of the selected features
print("First few rows of the selected features:")
X.head()
```

#### Applying k-means clustering

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=5, random_state=42)
kmeans.fit(X)
# Add cluster labels to the dataset
ml_data_encoded['cluster'] = kmeans.labels_
```

#### Calculate Mean Values for Each Cluster

```python
# Group by cluster and calculate mean values for selected columns
cluster_summary = ml_data_encoded.groupby('cluster')[analysis_cols].mean()

print("Numeric Features Summary:")
cluster_summary[numeric_features]
```

```python
# Display the summary for one-hot encoded features
print("\nOne-Hot Encoded Features Summary:")
cluster_summary[trait_name_cols + trait_type_cols + tpe_name_cols + condition_cols + month_cols]
```

#### Analyze Non-Numeric Columns Separately
* To understand the distribution of varieties, locations, and other attributes within each cluster.

```python
non_numeric_columns = ['variety_name', 'crop_name', 'town', 'country', 'tpe_description', 'stressors']

# Group by cluster and get unique values for non-numeric columns
non_numeric_summary = {}
for col in non_numeric_columns:
    non_numeric_summary[col] = ml_data_encoded.groupby('cluster')[col].unique()

# Display the unique values for each non-numeric column
for col in non_numeric_columns:
    print(f"\nUnique {col} per cluster:")
    for cluster, values in non_numeric_summary[col].items():
        print(f"Cluster {cluster}: {values}")
```

* This code groups the data by cluster and retrieves the unique values for each non-numeric column.
* This helps us understand the distribution of varieties, locations, and other attributes within each cluster.


#### Creating a Comprehensive Recommendation Table
* Combine the numeric and non-numeric summaries to create a comprehensive recommendation table.
* This table is useful for making recommendations to breeders based on the comprehensive characteristics of each cluster.


```python
# Create a comprehensive recommendation table
recommendation_table = pd.DataFrame({
    'Cluster': cluster_summary.index,
    'Numeric Summary': [cluster_summary.loc[cluster, numeric_features].to_dict() for cluster in cluster_summary.index],
    'Varieties': non_numeric_summary['variety_name'],
    'Crops': non_numeric_summary['crop_name'],
    'Towns': non_numeric_summary['town'],
    'Countries': non_numeric_summary['country'],
    'TPE Descriptions': non_numeric_summary['tpe_description'],
    'Stressors': non_numeric_summary['stressors'],
    'Key Traits': [cluster_summary.loc[cluster, trait_name_cols + trait_type_cols].to_dict() for cluster in cluster_summary.index],
    'Key TPEs': [cluster_summary.loc[cluster, tpe_name_cols].to_dict() for cluster in cluster_summary.index],
    'Key Conditions': [cluster_summary.loc[cluster, condition_cols].to_dict() for cluster in cluster_summary.index],
    'Key Months': [cluster_summary.loc[cluster, month_cols].to_dict() for cluster in cluster_summary.index]
})
recommendation_table
```

### Explaining the Recommendations Table
##### What Determines the Groups (Clusters)?
When we use K-Means clustering, the algorithm groups crop varieties based on similarities in their traits and environmental conditions. Here's what was used to form these groups:

**1. Numeric Features (Environmental Conditions)**
These are the measurable environmental factors that describe the conditions where the crop varieties grow:

* Temperature: Average temperature of the location.
* Humidity: Average humidity of the location.
* Rainfall: Average rainfall of the location.
* Latitude (lat) and Longitude (lon): Geographic location of where the crop is grown.

These factors help the algorithm group varieties that thrive in similar environmental conditions.

**2. One-Hot Encoded Categorical Features (Traits and Conditions)**
These are the categorical traits and conditions that have been converted into a numerical format (0s and 1s) so that the algorithm can understand them:

* Trait Names: Specific traits of the crop varieties (e.g., drought resistance, heat tolerance).
* Trait Types: Types of traits (e.g., desired, positive).
* TPE Names: Target Population Environment names (e.g., arid, semi-arid).
* Conditions: Environmental conditions (e.g., drought, heat).
* Months: Planting or harvesting months.

These factors help the algorithm group varieties that have similar traits and are suited for similar conditions.

<h4>How the Grouping Works</h4>

**1. Data Preparation (preprocessing):**
We prepare the data by scaling the numeric features and one-hot encoding the categorical features. This ensures that all features contribute equally to the clustering process.

**2. K-Means Clustering:**
The K-Means algorithm looks at all the selected features (numeric and one-hot encoded) and groups the crop varieties into clusters based on their similarities.
It calculates the distance between each variety's features and the center of each cluster, then assigns each variety to the nearest cluster.


**3. Cluster Formation:**
Varieties with similar traits and environmental conditions are grouped together.
For example, varieties that are drought-resistant and suited for high-temperature environments will likely be in the same cluster.

<h5>Example </h5>
Suppose we have the following data for three crop varieties:

| Variety Name | Temperature | Humidity | Drought Resistance | Heat Tolerance |
|--------------|-------------|----------|--------------------|----------------|
| Mucuvea      | 25.5        | 60.2     | 1                  | 1              |
| Otela        | 25.0        | 62.0     | 1                  | 1              |
| Kapaala      | 20.0        | 70.0     | 0                  | 0              |


- Mucuvea and Otela have similar temperature, humidity, and traits (both are drought-resistant and heat-tolerant).
- Kapaala has different temperature, humidity, and traits (not drought-resistant or heat-tolerant).
- The K-Means algorithm will likely group Mucuvea and Otela into one cluster and Kapaala into another cluster because of their similarities and differences.

#### Why This Approach Achieves CropSync Objectives**
* **Grouping by Similarity:** Varieties with similar traits and environmental conditions are grouped together, making it easier to identify which varieties are suited for specific conditions.
- **Informed Recommendations:** By understanding the characteristics of each cluster, you can recommend varieties to breeders based on their specific needs and environmental conditions.
- **Efficient Decision-Making:** Breeders can quickly see which varieties are best suited for their target locations and conditions.

- Numeric Features: Environmental conditions like temperature, humidity, and rainfall.

- K-Means Clustering: Groups varieties based on similarities in their traits and environmental conditions.
