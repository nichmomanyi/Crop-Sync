### Data cleaning and preparation
##### The variety_name, town, country, and stressors columns contain lists as strings (e.g., ["Mucuvea(2015)", "otela(2015)"]). We’ll convert these strings into proper Python lists for easier analysis.

```python
import ast

# Convert string representations of lists to actual lists
for col in ['variety_name', 'town', 'country', 'stressors']:
    cluster_full_summary[col] = cluster_full_summary[col].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else [])

cluster_full_summary.head()
```
